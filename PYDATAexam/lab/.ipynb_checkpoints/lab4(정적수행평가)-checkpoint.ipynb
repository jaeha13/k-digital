```python
import requests
from bs4 import BeautifulSoup
import re


comics_title = []
comics_summary = []
comics_ratings = [] 
for i in range(1,11):
    req = requests.get('https://comic.naver.com/genre/bestChallenge?&page=' + str(i))
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.select('#content > div > table > tr > td > div.challengeInfo > h6 > a')
    summary = soup.select('#content > div > table > tr > td > div.challengeInfo > div.summary')
    ratings = soup.select('#content > div > table > tr > td > div.challengeInfo > div.rating_type > strong')
    for dom in titles:
        comics_title.append(dom.text.strip())
    for dom in summary:
        comics_summary.append(dom.text)
    for dom in ratings:
        comics_ratings.append(dom.text)

print(comics_title)
print(comics_summary)
print(comics_ratings)
```
