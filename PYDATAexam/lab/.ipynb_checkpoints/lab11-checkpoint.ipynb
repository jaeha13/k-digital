```python
import pandas as pd

score = pd.read_csv('../data/score.csv')
print('컬럼명 : ', score.columns)
print('인덱스 : ', score.index)
display(score)
```

```python
score_n = pd.read_csv('../data/score_noheader.csv', header=None)
display(score_n)
print('컬럼명 : ', score_n.columns)
print('인덱스 : ', score_n.index)
print('컬럼명 추가 ------------------------------------')
score_n.columns = ['name', 'kor', 'eng', 'mat']
print('컬럼명 : ', score_n.columns)
print(score_n)
score_n['total'] = score_n['kor'] + score_n['eng'] + score_n['mat']
score_n['avg'] = round(score_n['total'] / 3, 2)
print('total과 avg열 추가 -----------------------------')
print(score_n)
```

```python
score_h = pd.read_csv('../data/score_header.csv', skiprows=3)
display(score_h)
print('컬럼명 : ', score_h.columns)
print('인덱스 : ', score_h.index)
score_h['total'] = score_h['kor'] + score_h['eng'] + score_h['mat']
score_h['avg'] = round(score_n['total'] / 3, 2)
display(score_h)
for _ in range(3):
    score_h.to_csv('../output/result.csv', mode='a', header=False, index=False)
```

```python
score2 = pd.read_excel('../data/score.xlsx')
display(score2)
score2['total'] = score2['kor'] + score2['eng'] + score2['mat']
score2['avg'] = round(score2['total'] / 3, 2)
score2.sort_values(by='total')
display(score2)
score2.to_excel('../output/score_result1.xlsx')
score2.to_excel('../output/score_result2.xlsx', index=False)
```
