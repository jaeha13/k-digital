```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
cctv = pd.read_csv("../data/cctv_seoul.csv")

cctv.set_index('구별', inplace=True)
```

```python
import seaborn as sns
mycolors = sns.color_palette('hls', len(cctv['CCTV수']))

plt.title('시각화 과제(1) 각 구의 CCTV 설치 현황', size=20)
plt.barh(cctv.index, cctv['CCTV수'], color=mycolors)
plt.grid(True)
plt.xlabel('각 구에 설치된 CCTV 댓수', size=15)
plt.ylabel('구이름', size=15)
plt.savefig("../output/hw5.png") 
plt.show()
```

```python
mycolors1 = sns.color_palette('winter', len(cctv.index))

plt.title('시각화 과제(2) 각 구의 CCTV 설치 현황', size=20)
cctv['CCTV수'].plot(kind='bar', grid=True, color=mycolors1)
plt.xlabel('구이름', size=15)
plt.xticks(rotation=45)
plt.ylabel('각 구에 설치된 CCTV 댓수', size=15)
plt.savefig("../output/hw6.png") 
plt.show()
```

```python
cctv['CCTV비율'] = cctv['CCTV수'] / cctv['인구수'] * 100
cctv['노인비율'] = cctv['고령자'] / cctv['인구수'] * 100
display(cctv.head())
```

```python
mycolors2 = sns.color_palette('terrain', len(cctv.index))

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.title('각 구의 인구수 대비 CCTV 비율')
plt.bar(cctv.index, cctv['CCTV비율'], color=mycolors2)
plt.ylabel('CCTV비율', size=8)
plt.subplot(2, 1, 2)
plt.title('각 구의 인구수 대비 고령자 비율')
plt.ylabel('노인비율', size=8)
plt.plot(cctv.index, cctv['노인비율'], ls='-', color='yellow')
plt.scatter(cctv.index, cctv['노인비율'], linewidths=0.5, color=mycolors2)
plt.winter()

plt.tight_layout()
plt.savefig('../output/hw7.png')
```
