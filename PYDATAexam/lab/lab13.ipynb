```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
font_path = "../data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

data = {
    '1분기' : [500, 690, 1100, 1500, 1990, 1020],
    '2분기' : [450, 700, 1030, 1650, 2020, 1600],
    '3분기' : [520, 820, 1200, 1700, 2300, 2200],
    '4분기' : [610, 900, 1380, 1850, 2420, 2550],
}
index_name = [2015, 2016, 2017, 2018, 2019, 2020]
df = pd.DataFrame(data, index=index_name)
display(df)
```

```python
import numpy as np

plt.figure(figsize=(10, 6))
plt.title('시각화 과제1', color='purple')
plt.plot(df)
plt.legend(df.columns, loc = 0)
plt.xticks(df.index, labels=[str(i) + '년' for i in df.index])
plt.savefig('../output/hw1.png')
plt.show()
```

```python
df1 = df.transpose()

plt.figure(figsize=(10, 6))
plt.title('시각화 과제2', color='red')
plt.plot(df1, label=[str(i) + '년' for i in df1.columns])
plt.legend(loc=0)
plt.savefig('../output/hw2.png')
plt.show()
```

```python
q_sum = []
for i in df.columns:
    q_sum_i = 0
    for j in df.index:
        q_sum_i += df[i][j]
    q_sum.append(q_sum_i)

total = sum(q_sum)

plt.figure(figsize=(10, 6))
explode = [0.05, 0.05, 0.05, 0.05]
plt.pie(q_sum / total, labels=df.columns, autopct='%.1f%%', startangle=0, counterclock=True, 
        explode=explode, shadow=True, colors = ['red', 'green', 'firebrick', 'purple'])
plt.savefig('../output/hw3.png')
plt.show()
```

```python
q_sum1 = []
for i in df1.columns:
    q_sum_i = 0
    for j in df1.index:
        q_sum_i += df1[i][j]
    q_sum1.append(q_sum_i)

total1 = sum(q_sum1)

plt.figure(figsize=(10, 6))
plt.pie(q_sum1 / total1, labels=[str(i) + '년' for i in df1.columns], 
        autopct='%.1f%%', startangle=0,  counterclock=True,
       colors=['red', 'green', 'firebrick', 'purple', 'yellow', 'pink'])
plt.savefig('../output/hw4.png')
plt.show()
```
