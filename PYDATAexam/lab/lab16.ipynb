```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
## 1
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
#1 nan이 있는 행 삭제
display(iris_x.head(10).dropna(subset=iris_x.columns,axis=0))
```

```python
#2 nan이 있는 두개 이상 존재하는 행 삭제
display(iris_x.dropna(subset=iris_x.columns,thresh=2,axis=0).head(10))
```

```python
#3 'sepal_length','sepal_width' 에 nan 이 존재하는 행 삭제
display(iris_x.dropna(subset=['sepal_length','sepal_width'],axis=0).head(10))
```

```python
#4 모든 nan을 0으로 채움
display(iris_x.fillna(0).head(10))
```

```python
#5 컬럼별 평균값으로 채우시
print(iris_x.mean(axis=0))
display(iris_x.fillna(iris_x.mean(axis=0).round(1)).head(10))
```

```python
#6
display(iris_x.fillna(10))
```

```python
#7
display(iris_x.fillna(method='ffill'))
```
